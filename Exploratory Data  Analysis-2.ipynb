### Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

```python
import matplotlib.pyplot as plt

# Create histogram of flight prices
plt.figure(figsize=(10, 6))
plt.hist(flight_data['Price'], bins=20, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.grid(True)
plt.show()
```

### Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
# Calculate range of prices
min_price = flight_data['Price'].min()
max_price = flight_data['Price'].max()

print(f"Minimum Price: {min_price}")
print(f"Maximum Price: {max_price}")
```

### Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.

```python
import seaborn as sns

# Create boxplot of flight prices by airline
plt.figure(figsize=(12, 8))
sns.boxplot(x='Airline', y='Price', data=flight_data)
plt.xlabel('Airline')
plt.ylabel('Price')
plt.title('Flight Prices by Airline')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Q5. Are there any outliers in the dataset? Identify any potential outliers using a boxplot and describe how they may impact your analysis.

```python
# Identify outliers using boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(flight_data['Price'])
plt.xlabel('Flight Prices')
plt.title('Identifying Outliers in Flight Prices')
plt.grid(True)
plt.show()

# Describe potential impact of outliers
outliers = flight_data[flight_data['Price'] > 1.5 * (flight_data['Price'].quantile(0.75) - flight_data['Price'].quantile(0.25))]
print("Potential outliers:")
print(outliers)
```

### Q12. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize the relationship.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Google Playstore dataset
playstore_data = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Clean and convert 'Installs' column to numeric
# Remove non-numeric characters and handle special cases
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',', '').str.replace('+', '').str.replace('Free', '0').str.replace('Paid', '0').astype(float)

# Create scatter plot of installs vs. rating
plt.figure(figsize=(10, 6))
plt.scatter(playstore_data['Installs'], playstore_data['Rating'], alpha=0.5)
plt.title('App Installs vs. Rating')
plt.xlabel('Installs')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Google Playstore dataset
playstore_data = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Convert 'Installs' column to numeric (handling 'Free', 'Paid', etc.)
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',', '').str.replace('+', '').str.replace('Free', '0').str.replace('Paid', '0').astype(float)

# Convert 'Content Rating' column to numerical labels
# Assuming mapping 'Everyone' to 0, 'Teen' to 1, 'Mature 17+' to 2, etc.
content_rating_mapping = {
    'Everyone': 0,
    'Everyone 10+': 1,
    'Teen': 2,
    'Mature 17+': 3,
    'Adults only 18+': 4,
    'Unrated': 5
}
playstore_data['Content Rating'] = playstore_data['Content Rating'].map(content_rating_mapping)

# Create scatter plot of content rating vs. rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Content Rating', y='Rating', data=playstore_data)
plt.title('App Content Rating vs. Rating')
plt.xlabel('Content Rating')
plt.ylabel('Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
import pandas as pd

# Load the Google Playstore dataset
playstore_data = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Clean and convert 'Installs' column to numeric
# Remove non-numeric characters and handle special cases
playstore_data['Installs'] = playstore_data['Installs'].str.replace(',', '').str.replace('+', '').str.replace('Free', '0').str.replace('Paid', '0').replace('Varies with device', None).astype(float)

# Find top 10 most popular apps by number of installs
top_10_apps = playstore_data.nlargest(10, 'Installs')[['App', 'Installs']]

print(top_10_apps)
```
