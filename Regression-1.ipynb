

**Q1. Explain the difference between simple linear regression and multiple linear regression. Provide an example of each.**

- **Simple Linear Regression:** In simple linear regression, there is a single independent variable \( X \) that predicts the dependent variable \( Y \). The relationship between \( X \) and \( Y \) is assumed to be linear, and the model equation is \( Y = \beta_0 + \beta_1 X + \epsilon \), where \( \beta_0 \) is the intercept, \( \beta_1 \) is the slope coefficient, and \( \epsilon \) is the error term.

  Example: Predicting house prices based on the square footage of the house.

- **Multiple Linear Regression:** In multiple linear regression, there are multiple independent variables \( X_1, X_2, \ldots, X_p \) that predict the dependent variable \( Y \). The model equation extends to \( Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \ldots + \beta_p X_p + \epsilon \).

  Example: Predicting a student's exam score based on study hours, previous exam scores, and attendance.

**Q2. Discuss the assumptions of linear regression. How can you check whether these assumptions hold in a given dataset?**

Linear regression assumes:
- Linearity: The relationship between independent and dependent variables should be linear.
- Independence: Residuals (errors) should be independent of each other.
- Homoscedasticity: Residuals should have constant variance.
- Normality: Residuals should be normally distributed.

To check these assumptions:
- **Residual plots:** Check for linearity and homoscedasticity.
- **Normality tests:** Test residuals for normality.
- **Durbin-Watson test:** Test for independence of residuals.

**Q3. How do you interpret the slope and intercept in a linear regression model? Provide an example using a real-world scenario.**

- **Intercept (\( \beta_0 \)):** It represents the predicted value of \( Y \) when all \( X \) variables are zero.
- **Slope (\( \beta_1 \)):** It represents the change in \( Y \) for a one-unit change in \( X \), holding other variables constant.

Example: In a linear regression model predicting salary based on years of experience, \( \beta_0 \) would be the estimated starting salary for someone with zero years of experience, and \( \beta_1 \) would be the estimated increase in salary for each additional year of experience.

**Q4. Explain the concept of gradient descent. How is it used in machine learning?**

Gradient descent is an optimization algorithm used to minimize the loss function of a machine learning model. It works by iteratively adjusting the model parameters (weights) in the direction that reduces the gradient of the loss function. This process continues until the algorithm converges to a minimum, ideally the global minimum, of the loss function.

**Q5. Describe the multiple linear regression model. How does it differ from simple linear regression?**

Multiple linear regression extends simple linear regression by incorporating multiple independent variables to predict a single dependent variable. The model equation includes multiple slope coefficients (\( \beta_1, \beta_2, \ldots, \beta_p \)) corresponding to each independent variable.

**Q6. Explain the concept of multicollinearity in multiple linear regression. How can you detect and address this issue?**

Multicollinearity occurs when independent variables in a regression model are highly correlated with each other. It can cause issues in the interpretation of coefficients and lead to unstable estimates.

Detection methods:
- **Correlation matrix:** Check pairwise correlations between independent variables.
- **Variance Inflation Factor (VIF):** Quantifies the severity of multicollinearity.

Addressing methods:
- **Feature selection:** Remove one of the correlated variables.
- **Principal Component Analysis (PCA):** Transform correlated variables into orthogonal components.

**Q7. Describe the polynomial regression model. How is it different from linear regression?**

Polynomial regression models the relationship between the independent variable \( X \) and the dependent variable \( Y \) as an \( n \)-th degree polynomial. This allows for more flexibility than simple linear regression, which assumes a linear relationship between \( X \) and \( Y \).

**Q8. What are the advantages and disadvantages of polynomial regression compared to linear regression? In what situations would you prefer to use polynomial regression?**

Advantages:
- **Flexibility:** Can capture non-linear relationships between variables.
- **Higher accuracy:** Can fit more complex data patterns.

Disadvantages:
- **Overfitting:** Higher-degree polynomials can lead to overfitting.
- **Interpretability:** Interpretation of coefficients becomes more complex.

Use cases:
- **Curvilinear data:** When the relationship between variables is not linear.
- **Exploratory analysis:** To understand complex relationships before applying more sophisticated models.
