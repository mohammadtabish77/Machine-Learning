

**Q1. Pearson correlation coefficient between studying time and exam scores:**

The Pearson correlation coefficient measures the linear relationship between two continuous variables. Let's denote:
- \( X \): Time spent studying for the exam.
- \( Y \): Final exam scores.

Assume you have collected data for these variables across a sample of students. The formula for Pearson correlation coefficient \( r \) is given by:

\[ r = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum (X_i - \bar{X})^2 \sum (Y_i - \bar{Y})^2}} \]

where \( \bar{X} \) and \( \bar{Y} \) are the means of \( X \) and \( Y \) respectively, and \( (X_i, Y_i) \) are data points.

- **Interpretation:**
  - \( r > 0 \): Indicates a positive linear relationship; as studying time increases, exam scores tend to increase.
  - \( r < 0 \): Indicates a negative linear relationship; as studying time increases, exam scores tend to decrease.
  - \( r = 0 \): Indicates no linear relationship between studying time and exam scores.

**Q2. Spearman's rank correlation between sleep and job satisfaction:**

Spearman's rank correlation coefficient assesses the strength and direction of monotonic association between two ranked variables. Let's denote:
- \( X \): Amount of sleep individuals get each night (in hours).
- \( Y \): Overall job satisfaction level (ranked from 1 to 10).

Assume you have collected ranked data for these variables across individuals. The formula for Spearman's rank correlation coefficient \( \rho \) is:

\[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \]

where \( d_i \) is the difference between the ranks of corresponding variables \( X \) and \( Y \), and \( n \) is the number of observations.

- **Interpretation:**
  - \( \rho = 1 \): Perfect monotonic relationship (increasing).
  - \( \rho = -1 \): Perfect monotonic relationship (decreasing).
  - \( \rho = 0 \): No monotonic relationship.

**Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.**

To calculate the Pearson correlation coefficient between the number of sales calls made per day and the number of sales made per week for a sample of 30 sales representatives, we can follow these steps:

Let's denote:
- \( X \): Number of sales calls made per day.
- \( Y \): Number of sales made per week.

Assume you have collected numerical data for these variables across the sample of 30 sales representatives.

### Steps to Calculate Pearson Correlation Coefficient \( r \):

1. **Compute the Means:**
   Calculate the mean (\( \bar{X} \)) and (\( \bar{Y} \)) of \( X \) and \( Y \) respectively.
   \[ \bar{X} = \frac{\sum X_i}{n} \]
   \[ \bar{Y} = \frac{\sum Y_i}{n} \]
   where \( X_i \) and \( Y_i \) are individual data points, and \( n \) is the number of observations (30 in this case).

2. **Compute the Covariance:**
   Calculate the covariance (\( \text{cov}(X, Y) \)) between \( X \) and \( Y \).
   \[ \text{cov}(X, Y) = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{n-1} \]

3. **Compute the Standard Deviations:**
   Calculate the standard deviations (\( s_X \) and \( s_Y \)) of \( X \) and \( Y \) respectively.
   \[ s_X = \sqrt{\frac{\sum (X_i - \bar{X})^2}{n-1}} \]
   \[ s_Y = \sqrt{\frac{\sum (Y_i - \bar{Y})^2}{n-1}} \]

4. **Calculate the Pearson Correlation Coefficient \( r \):**
   \[ r = \frac{\text{cov}(X, Y)}{s_X \cdot s_Y} \]

### Interpretation of \( r \):

- \( r > 0 \): Indicates a positive linear relationship; as the number of sales calls per day increases, the number of sales made per week tends to increase.
- \( r < 0 \): Indicates a negative linear relationship; as the number of sales calls per day increases, the number of sales made per week tends to decrease.
- \( r = 0 \): Indicates no linear relationship between the number of sales calls per day and the number of sales made per week.
